```python
import numpy as np
import pandas as pd
from getStockPrices import get_all_data
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
get_all_data()
```

```python
with open('stock_data.json','r') as f:
    data = json.load(f)

df = pd.DataFrame(data)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df
```

```python
# df['date'] = pd.to_datetime(df['date'])
df["price"] = df["price"].str.replace(',', '')
df['price'] = pd.to_numeric(df['price'])
df["change"] = df["change"].str.replace(',', '')
df["change"] = pd.to_numeric(df["change"])
df["changePercent"] = df["changePercent"].str.replace('%', '').str.replace('(', '').str.replace(')', '')
df['changePercent'] = pd.to_numeric(df['changePercent'])
```

```python
df.info()
```

Bar Chart of Stock Prices

```python
fig, ax = plt.subplots(figsize=(10,5))
bar_df = df.drop(15)
sns.barplot(x='symbol', y='price', data=bar_df, ax=ax)
ax.set_title('Stock Prices')
ax.set_ylabel('Price')
ax.set_xlabel('Symbol')
plt.show()
```

Area Chart of Stock Prices

```python
fig, ax = plt.subplots(figsize=(10,5))
area_df = df.drop(15)
sns.lineplot(x='symbol', y='price', data=bar_df, ax=ax)
ax.set_title('Stock Prices')
ax.set_ylabel('Price')
ax.set_xlabel('Symbol')
plt.show()
```

Scatter Plot

```python
fig, ax = plt.subplots(figsize=(10,5))
scatter_df = df.drop(15)
sns.scatterplot(x='symbol', y='price', data=scatter_df, ax=ax)
ax.set_title('Stock Prices')
ax.set_ylabel('Price')
ax.set_xlabel('Date')
plt.show()
```

Box Plot

```python
fig, ax = plt.subplots(figsize=(10,5))
box_df = df.drop(15)
sns.boxplot(x='symbol', y='price', data=box_df, ax=ax)
ax.set_title('Stock Prices')
ax.set_ylabel('Price')
ax.set_xlabel('Symbol')
plt.show()
```

Violin Plot

```python
fig, ax = plt.subplots(figsize=(10,5))
violin_df = df.drop(15)
sns.violinplot(x='symbol', y='price', data=violin_df, ax=ax)
ax.set_title('Stock Prices')
ax.set_ylabel('Price')
ax.set_xlabel('Symbol')
plt.show()
```

Histogram

```python
fig, ax = plt.subplots(figsize=(10,5))
hist_df = df.drop(15)
sns.histplot(hist_df['price'], ax=ax)
ax.set_title('Stock Prices')
ax.set_ylabel('Price')
ax.set_xlabel('Symbol')
plt.show()
```

Plot the biggest changes in price

```python
fig, ax = plt.subplots(figsize=(10,5))

change_df = df.sort_values(by='change', ascending=False)
change_df = change_df.drop(15)
sns.barplot(x='symbol', y='change', data=change_df, ax=ax)
ax.set_title('Stock Prices')
ax.set_ylabel('Price')
ax.set_xlabel('Symbol')
plt.show()

```
