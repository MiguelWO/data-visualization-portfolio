Este dataset es relevante porque nos permite obtener una visión general de la distribución geográfica de los países, así como explorar las disparidades económicas y sociales entre ellos.

Razón de la elección del dataset:

El dataset fue seleccionado debido a su potencial para analizar y comprender mejor la diversidad de países en términos de sus características económicas y sociales. Al examinar las variables principales presentes en el dataset, como la región, la población, el área geográfica, la densidad de población, la línea costera, la migración neta, la mortalidad infantil, el PIB per cápita y la tasa de alfabetización, podemos obtener una imagen más clara de cómo estos factores están distribuidos en diferentes regiones del mundo.

Variables principales:

Región: Esta variable indica la región geográfica a la que pertenece cada país. Puede incluir categorías como América del Norte, América del Sur, Europa, Asia, África, entre otras. El análisis de esta variable nos permite examinar las diferencias y similitudes entre las regiones en términos de las variables económicas y sociales restantes.

Población: Esta variable representa la cantidad total de habitantes en cada país. Proporciona información importante sobre la dimensión demográfica de cada nación y puede ser utilizada para comparar la magnitud relativa de las poblaciones entre diferentes países.

Área (sq. mi.): Esta variable indica el área geográfica de cada país en millas cuadradas. Al considerar esta variable en conjunto con la población, podemos obtener una idea de la densidad de población y la distribución espacial de los países.

Densidad de población (por sq. mi.): Esta variable calcula la densidad de población al dividir la población total del país por su área geográfica. Nos ayuda a entender cómo se distribuye la población en diferentes países y a comparar la concentración de personas en diferentes regiones.

Línea costera (relación costa/área): Esta variable representa la relación entre la longitud total de la línea costera de un país y su área terrestre. Migración neta: Esta variable indica el saldo neto de migración (inmigración menos emigración) en un país determinado. Nos permite comprender si un país está experimentando un flujo neto de migración positivo o negativo, lo que puede tener implicaciones significativas en términos de cambios demográficos y socioeconómicos.

# 2. Análisis exploratorio de datos

```python
# Importar librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar dataset
df = pd.read_csv('countries-of-the-world.csv', decimal=',')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
# Distribuciones de las variables numéricas
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
numeric_columns
```

```python
# Histogramas de las variables numéricas
for col in numeric_columns:
    plt.figure()
    sns.histplot(data=df, x=col, bins=20)
    plt.title(col)
```

```python
# Distribuciones de las variables categóricas
categorical_columns = df.select_dtypes(include='object').columns.tolist()
categorical_columns
```

```python
# Diagramas de barras de las variables categóricas
for col in categorical_columns:
    plt.figure()
    sns.countplot(data=df, x=col)
    plt.xticks(rotation=90)
    plt.title(col)
```

# 3. Visualización de datos

```python
# Histograma de la variable 'GDP ($ per capita)
sns.histplot(data=df, x='GDP ($ per capita)', bins=20)
```

```python
# Diagrama de cajas de la variable 'Infant mortality (per 1000 births)'
sns.boxplot(data=df, x='Infant mortality (per 1000 births)')
plt.title('Infant mortality (per 1000 births)')
```

```python
# Diagrama de barras de la variable 'Region'
sns.countplot(data=df, x='Region')
plt.xticks(rotation=90)
plt.title('Region')
```

```python
# Gráfico de dispersión de 'Literacy (%)' vs 'GDP ($ per capita)'
sns.scatterplot(data=df, x='Literacy (%)', y='GDP ($ per capita)')
plt.title('Literacy (%) vs GDP ($ per capita)')
```

```python
# Comparar el PIB per cápita de los países en diferentes regiones
avg_gdp = df.groupby('Region')['GDP ($ per capita)'].mean().sort_values(ascending=False)
avg_gdp
```

```python
avg_gdp.plot(kind='bar')
plt.xticks(rotation=90)
plt.title('Average GDP per capita by Region')
```

```python
# Violin plot de 'Region' vs 'GDP ($ per capita)'
sns.violinplot(data=df, x='Region', y='GDP ($ per capita)')
plt.xticks(rotation=90)
plt.title('Region vs GDP ($ per capita)')
```

```python
# Diagrama de barras de 'Region' vs 'GDP ($ per capita)'
sns.barplot(data=df, x='Region', y='GDP ($ per capita)')
plt.xticks(rotation=90)
plt.title('Region vs GDP ($ per capita)')
```

# 4. Correlación entre variables

```python
corr = df.corr()
corr
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(corr, annot=True)
plt.title('Correlation matrix')
```
